```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec

# 1. Define the Data
x_labels = ['2005-2009\nn = 7', '2010-2014\nn = 102', '2015-2019\nn = 183', '2020-2025\nn = 447']

# Group A: Study Design (7 rows)
data_a = np.array([
    [29, 45, 46, 51],
    [29, 61, 52, 56],
    [43, 56, 52, 54],
    [14, 30, 40, 43],
    [43, 10,  9,  8],
    [ 0, 10, 15, 18],
    [14, 14, 14, 16]
])
ylabels_a = ['Survey/Questionnaire', 'Qualitative/Interview', 'Mixed Methods',
             'Statistical/Quantitative', 'Climate Modelling', 'Remote Sensing/GIS',
             'Review/Meta-analysis']

# Group B: Hazard Type (8 rows)
data_b = np.array([
    [43, 52, 49, 53],
    [71, 39, 54, 49],
    [29, 39, 34, 36],
    [71, 65, 69, 71],
    [29, 33, 37, 36],
    [43, 32, 30, 29],
    [43, 25, 26, 29],
    [43, 37, 39, 41]
])
ylabels_b = ['Cyclone/Storm Surge', 'Salinity Intrusion', 'Flood', 'Sea-Level Rise',
             'Drought', 'Heat Stress', 'Erosion', 'Rainfall Variability']

# Group C: Adaptation Family (8 rows)
data_c = np.array([
    [29, 25, 32, 36],
    [57, 36, 34, 44],
    [29, 54, 60, 60],
    [43, 47, 48, 50],
    [ 0, 23, 33, 34],
    [29, 16, 21, 23],
    [29, 13, 15, 20],
    [14, 25, 19, 28]
])
ylabels_c = ['Water Management', 'Institutional & Policy', 'Social & Community',
             'Infrastructure', 'Agronomic', 'Financial', 'Ecosystem based', 'Migration']

# 2. Set up the Figure and Grid
# We use height ratios 7:8:8 so all cells remain perfectly square/uniform in size
fig = plt.figure(figsize=(12, 14))
gs = GridSpec(3, 2, width_ratios=[18, 0.50], height_ratios=[7, 8, 8], wspace=0.05, hspace=0.05)

# Create axes for heatmaps
ax_a = fig.add_subplot(gs[0, 0])
ax_b = fig.add_subplot(gs[1, 0])
ax_c = fig.add_subplot(gs[2, 0])

# Create axes for colorbars
cax_a = fig.add_subplot(gs[0, 1])
cax_b = fig.add_subplot(gs[1, 1])
cax_c = fig.add_subplot(gs[2, 1])

# 3. Plot the Heatmaps
# Section A
sns.heatmap(data_a, ax=ax_a, cmap='Blues', annot=True, fmt='d', cbar_ax=cax_a,
            vmin=0, vmax=100, yticklabels=ylabels_a, xticklabels=False,
            linecolor='white', linewidths=0.5, annot_kws={"fontsize": 15, "fontweight": "bold"})
cax_a.tick_params(labelsize=12)
ax_a.set_ylabel('(A) STUDY DESIGN (%)', fontweight='bold', labelpad=15)

# Section B
sns.heatmap(data_b, ax=ax_b, cmap='Reds', annot=True, fmt='d', cbar_ax=cax_b,
            vmin=0, vmax=100, yticklabels=ylabels_b, xticklabels=False,
            linecolor='white', linewidths=0.5, annot_kws={"fontsize": 16, "fontweight": "bold"})
cax_b.tick_params(labelsize=12)
ax_b.set_ylabel('(B) HAZARD TYPE (%)', fontweight='bold', labelpad=15)

# Section C
sns.heatmap(data_c, ax=ax_c, cmap='Greens', annot=True, fmt='d', cbar_ax=cax_c,
            vmin=0, vmax=100, yticklabels=ylabels_c, xticklabels=x_labels,
            linecolor='white', linewidths=0.5, annot_kws={"fontsize": 16, "fontweight": "bold"})
cax_c.tick_params(labelsize=12)
ax_c.tick_params(axis='x', labelsize=12)
ax_c.set_ylabel('(C) ADAPTATION FAMILY (%)', fontweight='bold', labelpad=15)

# 4. Formatting and Clean Up
# Make y-axis labels horizontal and aligned right
for ax in [ax_a, ax_b, ax_c]:
    ax.tick_params(axis='y', labelrotation=0)

# Reduce color bar height by 30%
for cax in [cax_a, cax_b, cax_c]:
    pos = cax.get_position()
    new_height = pos.height * 0.70 # Reduce by 30%
    y_center = pos.ymin + pos.height / 2  # Corrected from pos.y to pos.ymin
    new_bottom = y_center - new_height / 2
    cax.set_position([pos.x0, new_bottom, pos.width, new_height]) # Changed pos.x to pos.x0 for consistency

# Title
fig.suptitle('CONSOLIDATED ANALYSIS OF ENVIRONMENTAL ADAPTATION LITERATURE TRENDS (2005-2025)',
             fontsize=16, fontweight='bold', y=0.92)
plt.savefig('COnsolidated_heatmap.png', bbox_inches='tight', dpi=500)
plt.savefig('COnsolidated_heatmap.pdf', bbox_inches='tight')
plt.show()
```
